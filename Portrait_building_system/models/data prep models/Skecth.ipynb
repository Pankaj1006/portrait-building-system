```
def dodge(front,back): 
    result=front*255/(255-back)  
    result[result>255]=255 
    result[back==255]=255 
    return result.astype('uint8')
```

```
def sketch(img):
    start_img = imageio.imread(img)
    gray_img = np.dot(start_img[...,:3], [0.299, 0.587, 0.114])
    inverted_img = 255 - gray_img
    blur_img = scipy.ndimage.filters.gaussian_filter(inverted_img,sigma=10)
    final_img= dodge(blur_img,gray_img)
    return final_img
```

```
i=1
for img in glob.glob("/content/drive/My Drive/face1/*.jpg"):
  image = sketch(img)
  path = "/content/drive/My Drive/face/"
  fileName="image"+str(i)+".jpg"
  cv2.imwrite(os.path.join(path , fileName), image)
  print('\r', i, end="")
  i+=1
```
